```python
# Lista de tarefas
tarefas = []

def listar_tarefas():
    print("Lista de Tarefas:")
    for idx, tarefa in enumerate(tarefas, 1):
        status = "[x]" if tarefa["realizada"] else "[ ]"
        print(f"{idx}. {tarefa['descricao']} {status}")

def registrar_tarefa():
    descricao = input("Digite a descrição da tarefa: ")
    # Garante que a descrição começa com maiúscula
    descricao = descricao.capitalize()
    nova_tarefa = {"descricao": descricao, "realizada": False}
    tarefas.append(nova_tarefa)
    print("Tarefa registrada!!!")

def marcar_como_realizada():
    listar_tarefas()
    identificador = int(input("Digite o ID da tarefa a ser marcada como realizada: "))
    if 1 <= identificador <= len(tarefas):
        tarefa = tarefas.pop(identificador - 1)
        tarefa["realizada"] = True
        tarefas.insert(0, tarefa)
        print("Tarefa marcada como realizada!")
    else:
        print("ID inválido ou tarefa já realizada.")

def editar_tarefa():
    listar_tarefas()
    identificador = int(input("Digite o ID da tarefa a ser editada: "))
    if 1 <= identificador <= len(tarefas):
        nova_descricao = input("Digite a nova descrição da tarefa: ")
        # Mantém o status da tarefa e o identificador
        tarefas[identificador - 1]["descricao"] = nova_descricao.capitalize()
        print("Tarefa editada!")
    else:
        print("ID inválido.")

# Exemplo de uso
registrar_tarefa()
listar_tarefas()
marcar_como_realizada()
listar_tarefas()
editar_tarefa()
listar_tarefas()
```

```python
import json

# Função para carregar as tarefas do arquivo
def carregar_tarefas():
    try:
        with open("tarefas.json", "r") as arquivo:
            return json.load(arquivo)
    except FileNotFoundError:
        return []

# Função para salvar as tarefas no arquivo
def salvar_tarefas(tarefas):
    with open("tarefas.json", "w") as arquivo:
        json.dump(tarefas, arquivo)

# Inicializa a lista de tarefas carregando do arquivo
tarefas = carregar_tarefas()

# Restante do código permanece o mesmo

def listar_tarefas():
    print("Lista de Tarefas:")
    for idx, tarefa in enumerate(tarefas, 1):
        status = "[x]" if tarefa["realizada"] else "[ ]"
        print(f"{idx}. {tarefa['descricao']} {status}")

def registrar_tarefa():
    descricao = input("Digite a descrição da tarefa: ")
    descricao = descricao.capitalize()
    nova_tarefa = {"descricao": descricao, "realizada": False}
    tarefas.append(nova_tarefa)
    salvar_tarefas(tarefas)  # Salva as tarefas após registrar uma nova

def marcar_como_realizada():
    listar_tarefas()
    identificador = int(input("Digite o ID da tarefa a ser marcada como realizada: "))
    if 1 <= identificador <= len(tarefas):
        tarefa = tarefas.pop(identificador - 1)
        tarefa["realizada"] = True
        tarefas.insert(0, tarefa)
        salvar_tarefas(tarefas)  # Salva as tarefas após marcar como realizada
        print("Tarefa marcada como realizada!")
    else:
        print("ID inválido ou tarefa já realizada.")

def editar_tarefa():
    listar_tarefas()
    identificador = int(input("Digite o ID da tarefa a ser editada: "))
    if 1 <= identificador <= len(tarefas):
        nova_descricao = input("Digite a nova descrição da tarefa: ")
        tarefas[identificador - 1]["descricao"] = nova_descricao.capitalize()
        salvar_tarefas(tarefas)  # Salva as tarefas após editar
        print("Tarefa editada!")
    else:
        print("ID inválido.")

# Exemplo de uso
registrar_tarefa()
listar_tarefas()
marcar_como_realizada()
listar_tarefas()
editar_tarefa()
listar_tarefas()
```
